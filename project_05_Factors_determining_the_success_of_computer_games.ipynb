Взглянем подробнее.

Пользователи оценивали игру от 0 до 10, максимальная данная оценка - 9.7. Видим, что помимо пропущенных значений, представленых `nan`, в таблице так же есть значение `tbd`. 

```python
df[df.user_score == 'tbd'].name.count()
```

2424 игры с такой пользовательской оценкой. Согласно [источнику](http://mobilradio.ru/information/vocabulary/tbd.htm), `tbd` может переводиться как "To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято." (с). Так как мы предполагаем, что заменять такие данные на средние, медианные или другие значения даже в зависимости от платформы, жанра и прочего будет некорректно, для нас такая пользовательская оценка ничем не отличается от пропуска в значениях `nan`.

Так же можно отметить тот факт, что критики оценивали игру по 100-балльной шкале, а пользователи по десятибальной. Хотелось бы привести их к единой шкале, поэтому превратим пользовательскую оценку в 100-балльную, умножив пользовательские оценки на 10. Так мы сможем по единой шкале оценить и мнения критиков, и мнения пользователей.

```python
df['user_score_normed'] = df.user_score.where(~((df.user_score == 'tbd') | (df.user_score.isna())))
df['user_score_normed'] = df['user_score_normed'].apply(float) * 10
```

Таким образом мы избавились от неопределенного значения пользовательской оценки, превратили данные в числовые и нормировали их на единую шкалу с оценкой критиков.

В таблице представлены следующие [рейтинги игр ESRB](https://www.esrb.org/ratings-guide/):
- `AO` - 'adults only' - для игроков 18+
- `E` - 'everyone' - для игроков любого возраста
- `E10+` - 'everyone 10+' - для игроков 10+
- `EC` - 'early childhood' - для игроков 3+
- `K-A` - 'kids to adults' - для игроков 6+, с 1998 заменен на рейтинг `E`
- `M` - 'mature' - для игроков 17+
- `RP` - 'rating pending' - рейтинг ожидает объявления ESRB
- `T` - 'teen' - для игроков 13+

В колонке присутствуют неявные дубликаты рейтинга: старый рейтинг `K-A` и более новый `E`. Заменим рейтинг на более новое обозначение. Так же в таблице есть значения с пока не присвоенным рейтингом, что для нашего исследования ничем не отличается от пропущенного значения. Объединим их в рейтинг `unknown`.

```python
df['rating'] = df['rating'].replace({'K-A': 'E', 'RP': 'unknown'})
df['rating'] = df['rating'].fillna('unknown')
```

### Как влияют на продажи внутри платформы отзывы пользователей и критиков?

Чтобы узнать, как влияют на продажи внутри платформ отзывы пользователей и критиков возьмем одну из платформ - `PS4` и построим для нее графики зависимости продаж от отзывов.

```python
# df_2014[df_2014.platform == 'PS3']
fig, axs = plt.subplots(1, 2, figsize=(14, 4))
ax1 = axs[0]
ax1.scatter(
    x = df_2014[df_2014.platform == 'PS4']['critic_score'], 
    y = df_2014[df_2014.platform == 'PS4']['total_sales'],
    alpha=0.5
)
ax1.set_xlabel("Оценка критиков (от 0 до 100)")
ax1.set_ylabel('Суммарная продажа, млн. проданных копий')

ax2 = axs[1]
ax2.scatter(
    x = df_2014[df_2014.platform == 'PS4']['user_score_normed'], 
    y = df_2014[df_2014.platform == 'PS4']['total_sales'],
    alpha=0.5
)
ax2.set_xlabel("Оценка пользователей (от 0 до 100)")
ax2.set_ylabel('Суммарная продажа, млн. проданных копий')

plt.suptitle('Зависимость продаж внутри платформы "PS4" от отзывов критиков и пользователей')
plt.show()
```

### Какие жанры являются самыми прибыльными? Выделяются ли жанры с высокими и низкими продажами?

Создадим сводную таблицу по суммарным продажам игр разных жанров по годам.

```python
pivot_genre = df_years.pivot_table(
    index=['genre', 'year_of_release'],
    values='total_sales',
    aggfunc='sum'
)
pivot_genre.columns = ['sum_total_sales']
pivot_genre.reset_index(inplace=True)
pivot_genre.head()
```

Посмотрим, какие жанры представлены в таблице.

```python
pivot_genre.genre.unique()
```

Для удобства построения графиков разобъем все имеющиеся жанры на два списка.

```python
genre_first = ['Action', 'Misc', 'Racing', 'Role-Playing', 'Shooter',  'Sports']
genre_second = ['Adventure', 'Fighting', 'Platform', 'Puzzle', 'Simulation', 'Strategy']
```

Построим <a id='genres_years'> график суммарных продаж по годам в зависимости от жанра игры </a>.

```python
fig, axs = plt.subplots(2, 1, figsize=(12, 8), constrained_layout=True)
ax1 = axs[0]
ax2 = axs[1]

for genre, data in pivot_genre[pivot_genre.genre.isin(genre_first)].groupby('genre'):
    data.plot(
        'year_of_release', 
        'sum_total_sales', 
        kind='line', 
        ylim=(0, 140), 
        label=genre, 
        ax=ax1, 
        style='-o',
        xlim=[1995, 2016], 
        grid=True, 
    )
    

ax1.set_ylabel('Суммарная продажа, млн. проданных копий')
ax1.set_xlabel('Жанр')


for genre, data in pivot_genre[pivot_genre.genre.isin(genre_second)].groupby('genre'):
    data.plot(
        'year_of_release',
        'sum_total_sales',
        kind='line',
        ylim=(0, 140),
        label=genre, 
        ax=ax2,
        style='-o',
        xlim=[1995, 2016],
        grid=True
    )
    
plt.suptitle('Распределение продаж всех платформ по годам')

ax2.set_ylabel('Суммарная продажа, млн. проданных копий')
ax2.set_xlabel('Год')

plt.show()
```

**Во все года**

## Проверка гипотез

В рамках данного исследования мы будем проверять две гипотезы:
1. Средние пользовательские рейтинги платформ `XOne` и `PC` одинаковые
2. Средние пользовательские рейтинги жанров `Action` и `Sports` разные.

### Гипотеза 1 
<a id='hyp1'> </a> **Средние пользовательские рейтинги платформ XOne и PC одинаковые**

Для проверки данной гипотезы применим гипотезу о равенстве средних двух генеральных совокупностей, используя имеющиеся выборки по актуальному периоду. 

Формулируем нулевую гипотезу как:

- **cредние пользовательские рейтинги платформ `XOne` и `PC` одинаковые**

Альтернативная гипотеза в данном случае будет следующей:

- **cредние пользовательские рейтинги платформ `XOne` и `PC` различаются**

Для того, чтобы понять, отвергнуть нам нулевую гипотезу или нет, применим  метод
`scipy.stats.ttest_ind(array1, array2, equal_var)`.
Параметры метода:
- `array1`, `array2` — массивы, содержащие выборки - в нашем случае это средние пользовательские рейтинги платформ `XOne` и `PC`;
- `equal_var` (от англ. equal variance, «равная дисперсия») — необязательный параметр, задающий, считать ли равными дисперсии генеральных совокупностей, из которых были сформированы выборки. 

По выборкам у нас пока нет значений дисперсии либо стандартного отклонения. Придется провести расчеты.

Так же необходимо задать **критический уровень статистической значимости** - вероятность отвергнуть верную нулевую гипотезу. Выберем значение в `5%`. Это значит, что если мы будем проводить аналогичные исследования 100 раз и проверять на основе имеющихся данных интересующую нас нулевую гипотезу, в 5 случаях из 100 мы отвергнем нулевую гипотезу, хотя она будет верной.

Задаем критический уровень статистической значимости.

```python
alpha = 0.05
```

Заводим переменные с выборками для пользовательских рейтингов игр разных платформ.

```python
array1 = df_2014[df_2014.platform == 'XOne']['user_score_normed'].dropna()
array2 = df_2014[df_2014.platform == 'PC']['user_score_normed'].dropna()
```

Посмотрим на статистику пользовательских отзывов на игры платформы `XOne`.

```python
array1.describe()
```

Рейтинг пользовательских отзывов на игры платформы `XOne` колеблется от 16 до 92 баллов, со средним значением в 66 баллов и медианой в 68 баллов. Стандартное отклонение 13.8.

Посмотрим на статистику пользовательских отзывов на игры платформы `PC`.

```python
array2.describe()
```

Рейтинг пользовательских отзывов на игры платформы `PC` колеблется от 14 до 93 баллов, со средним значением в 62 балла и медианой в 68 баллов. Стандартное отклонение 17.0.

Считаем дисперсии.


```python
print('Дисперсия выборки по платформе `XOne` равна', array1.std(ddof=1) ** 2)
print('Дисперсия выборки по платформе `PC` равна', array2.std(ddof=1) ** 2)
```

Дисперсии не равны, параметр `equal_var` при проверке гипотезы зададим как `False`.

Проверка статистической гипотезы.

```python
results = st.ttest_ind(array1, array2, equal_var=False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**

Не получилось отвергнуть нулевую гипотезу, принимаем гипотезу **cредние пользовательские рейтинги платформ `XOne` и `PC` равны**.
