Выяснить, какой тариф приносит больше денег, чтобы коммерческий департамент смог скорректировать рекламный бюджет.

## Описание тарифов

Нам предоставлено описание тарифов.

- **Тариф «Смарт»**
1. *Ежемесячная плата:* **550 рублей**
2. *Включено* **500 минут разговора, 50 сообщений и 15 Гб интернет-трафика**
3. *Стоимость услуг сверх тарифного пакета:* 
    - минута разговора: 3 рубля («Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута);    
    - сообщение: 3 рубля;    
    - 1 Гб интернет-трафика: 200 рублей.


- **Тариф «Ультра»**

1. *Ежемесячная плата:* **1950 рублей**
2. *Включено* **3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика**
3. *Стоимость услуг сверх тарифного пакета:* 
    - минута разговора: 1 рубль; 
    - сообщение: 1 рубль; 
    - 1 Гб интернет-трафика: 150 рублей.

**Важно:**
- «Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.
- Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.

- Так же мы предупреждены, что в данных есть звонки с нулевой продолжительностью - так обозначены пропущенные звонки.

В таблице `df_tariffs` представлена следующая информация о тарифах:

## Рассчеты минут, сообщений и интернет-трафика и анализ поведения клиентов

[Согласно принятому выше допущению](#assumption) мы будем считать, что **действие тарифа ограничено 30-дневным периодом**. 

Для подсчетов было решено добавить в каждую таблицу колонку с *днем от начала года* для дня подключения тарифа. А так же дату каждой операции пользователя перевести в день от начала года. Таким образом мы сможем сразу получить данные, *в какой день от начала использования тарифа* была совершена та или иная операция. 

*К примеру*, пользователь подключился в 145 день от начала года, а первую смс отправил в 146 день от начала года. Получаем 1 день от начала пользования тарифом (день подключения будем считать нулевым днем). Все операции до 30 дня мы будем считать первым периодом использования тарифа, с 30 до 60 дня - вторым и т.д.

```python
df_users[~(df_users['churn_date'].isna())].head()
```

```python
# день года подключения тарифа
df_users['first_day'] = df_users['reg_date'].dt.dayofyear
# день года прекращения использования тарифа или конца 2018 года
df_users['last_day'] = df_users['churn_date'].dt.dayofyear.where(~df_users['churn_date'].isna(), 365).astype('int16')
# продолжительность использования тарифа
df_users['duration'] = df_users['last_day'] - df_users['first_day']
# сколько полных периодов для пользователя уместилось в 2018 году
df_users['full_periods'] = df_users['duration'] // 30
```

```python
df_users.head()
```

В таблице мы рассчитали, сколько полных 30-дневных периодов было у пользователя - `full_periods`. 

Пользователи подключались в разные дни, а данные у нас ограничены 2018 годом. Таким образом для некоторых пользователей на конец 2018 года очередной 30-дневный период использования тарифа не был закончен, то есть за посдедний период для большинства пользователей у нас неполная информация. 

### Рассчеты: звонки и израсходованные минуты

Добавим в таблицу со звонками день от начала года совершения звонков пользователями. 

```python
df_calls['call_day'] = df_calls['call_date'].dt.dayofyear
```

Так же для дальнейших рассчетов добавим в таблицу со звонками информацию о первом дне использования тарифа пользователем и о самом тарифе.

```python
df_calls = df_calls.merge(
    df_users[['user_id', 'first_day', 'tariff', 'full_periods']], 
    how='left', 
    on='user_id')
```

Теперь в таблице есть все необходимые данные для рассчетов номера периода.

Рассчитаем порядковый номер 30-дневного периода использования тарифа пользователем. Для этого мы считаем разницу между днем совершения операции и днем подключения тарифа, далее целочисленно делим на количество дней в периоде, и чтобы период начинался с первого, а не с нулевого, прибавляем 1. 

Создадим в таблице со звонками колонку с номером периода, в который был произведен звонок.

```python
df_calls['call_period'] = (df_calls['call_day'] - df_calls['first_day']) // 30 + 1
```

```python
df_calls.head()
```

*Для большей достоверности* при рассчетах средних значений принято решение **учитывать все интервалы только для пользователей, отказавшихся от тарифа**. А для пользователей, продолжающих использование своего тарифа **на конец 2018 исключить из рассмотрения последний неполный период**. Конечно, они заплатили за него абонентскую плату и это может учитываться при общей выручке за 2018 год. Однако неполный период может занизить среднее количество операций в рамках тарифов для своих пользователей.

Создадим список отказавшихся от использования тарифов пользователей.

```python
refused_users = list(refused.user_id.unique())
```

Напишем функцию для выбора из всех периодов только либо периодов в 30 дней, либо всех периодов для отказавшихся пользователей.

```python
def considered_calls(data):
    '''Функция для учета всех периодов отказавшихся пользователей и только полных периодов не отказавшихся '''
    if data['user_id'] in refused_users:
        return True
    elif data['full_periods'] >= data['call_period']:
        return True
    return False
```

Применим функцию к таблице, чтобы получить колонку для корректных периодов.

```python
df_calls['full_30_days_or_refused'] = df_calls.apply(considered_calls, axis=1)
```

Оставим в таблице только корректные периоды.

```python
df_calls = df_calls[df_calls['full_30_days_or_refused'] == True]
```

Для рассчетов нам понадобится сводная таблица. Вынесем в неё необходимые данные: тариф, идентификационный номер пользователя, количество полных периодов и период самого звонка. Для рассчетов возьмем колонку с продолжительностью звонков и рассчитаем их сумму за период и количество совершенных звонков.

```python
calls_pivot = df_calls.pivot_table(
    index=['tariff', 'user_id', 'full_periods', 'call_period'], 
    values='duration', 
    aggfunc=['sum', 'count'])

calls_pivot.columns = ['sum_duration', 'count_duration']

calls_pivot = calls_pivot.reset_index()
```

Добавим в сводную таблицу информацию по тарифам.

```python
calls_pivot = calls_pivot.merge(df_tariffs[['tariff', 'minutes_included', 'rub_per_minute']], on='tariff', how='left')
```

```python
calls_pivot.head()
```

Подсчитаем, сколько дополнительных денег потратили пользователи за каждый период.

Для этого используем функцию `extra_money_minutes`, которая при использовании минут сверх тарифа умножает их на стоимость по тарифу.

```python
def extra_money_minutes(data):
    if data['minutes_included'] < data['sum_duration']:
        return int((data['sum_duration'] - data['minutes_included']) * data['rub_per_minute'])
    return 0
```

Добавим в таблицу колонку с дополнительными тратами за звонки.

```python
calls_pivot['extra_money_calls'] = calls_pivot.apply(extra_money_minutes, axis=1)    
calls_pivot[calls_pivot.extra_money_calls > 0].head()
```

Для рассчетов возьмем колонку с использованным интернет-трафиком и рассчитаем количество интернет-сессий и сумму потраченного трафика.

```python
internet_pivot = df_internet.pivot_table(
    index=['tariff', 'user_id', 'full_periods', 'session_period'], 
    values='mb_used', 
    aggfunc=['sum', 'count'])


internet_pivot.columns = ['sum_mb_used', 'count_sessions']
internet_pivot = internet_pivot.reset_index()
```

Подсчитаем, сколько дополнительных денег потратили пользователи за каждый период.

Для этого используем функцию `extra_money_internet`, которая при использовании трафика сверх тарифа умножает количество превышенных гигабайт их на стоимость по тарифу. 

Вспомним, что «Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.

```python
def extra_money_internet(data):
    if data['mb_per_month_included'] < data['sum_mb_used']:
        extra_mb = data['sum_mb_used'] - data['mb_per_month_included']
        extra_gb = np.ceil(extra_mb / data['mb_per_month_included'])
        extra_price = extra_gb * data['rub_per_gb']
        return int(extra_price)
    return 0
```

## Проверка гипотез

### Гипотеза 1
**средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются**

Для того, чтобы выяснить, отличается ли средняя выручка у пользователей разных тарифов, применим гипотезу о равенстве средних двух генеральных совокупностей, используя имеющуюся выборку по всем периодам пользователей двух тарифов. 

Формулируем нулевую гипотезу как:

- **средняя выручка пользователей тарифов «Ультра» и «Смарт» одинакова**.

Альтернативная гипотеза в данном случае будет следующей:

- **средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются**.

Для того, чтобы понять, отвергнуть нам нулевую гипотезу или нет, применим  метод
`scipy.stats.ttest_ind(array1, array2, equal_var)`.
Параметры метода:
- `array1`, `array2` — массивы, содержащие выборки - в нашем случае это выручка пользователей разных тарифов;
- `equal_var` (от англ. equal variance, «равная дисперсия») — необязательный параметр, задающий, считать ли равными дисперсии генеральных совокупностей, из которых были сформированы выборки. Так как по выборкам у нас есть значения стандартного отклонения, и они не равны, мы укажем этот параметр как `False`.

Так же необходимо задать **критический уровень статистической значимости** - вероятность отвергнуть верную нулевую гипотезу. Выберем значение в `5%`. Это значит, что если мы будем проводить аналогичные исследования 100 раз и проверять на основе имеющихся данных интересующую нас нулевую гипотезу, в 5 случаях из 100 мы отвергнем нулевую гипотезу, хотя она будет верной.

Так же для достоверности мы проверим каждую из гипотез два раза: 
- на выборке, учитывающей каждый полный период или каждый период для отказавшегося от тарифа пользователя,
- а так же на выборке со средними значениями выручки от каждого пользователя.

#### Проверка гипотезы 1 на выборке со всеми периодами

Задаем критический уровень статистической значимости.

```python
alpha = 0.05
```

Заводим переменные с выборками для разных тарифов.

```python
array1 = total_money_for_period[total_money_for_period['tariff'] == 'smart']['total_price']
array2 = total_money_for_period[total_money_for_period['tariff'] == 'ultra']['total_price']
```

Проверка статистической гипотезы.

```python
results = st.ttest_ind(array1, array2, equal_var=False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

- Подтвердить нулевую гипотезу не удалось, для всех периодов принимаем альтернативную гипотезу **средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются**.

#### Проверка гипотезы 1 на выборке со средними значениями по пользователям

Заводим переменные с выборками для разных тарифов.

```python
array1 = total_money_for_users[total_money_for_users['tariff'] == 'smart']['total_price']
array2 = total_money_for_users[total_money_for_users['tariff'] == 'ultra']['total_price']
```

Проверка статистической гипотезы.

```python
results = st.ttest_ind(array1, array2, equal_var=False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

- Подтвердить нулевую гипотезу также не удалось, для всех пользователей принимаем альтернативную гипотезу **средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются**.

### Гипотеза 2
**средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов**

По аналогии с гипотезой 1, формулируем нулевую гипотезу как:

- **средняя выручка пользователей из Москвы равна выручке пользователей из других регионов**.

Альтернативная гипотеза в данном случае будет следующей:

- **средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов**.

#### Проверка гипотезы 2 на выборке со всеми периодами

Заводим переменные с выборками для разных тарифов.

```python
array1 = total_money_for_period[total_money_for_period['city_category'] == 'Москва']['total_price']
array2 = total_money_for_period[total_money_for_period['city_category'] == 'другие']['total_price']
```

Проверка статистической гипотезы.

Было решено, опираясь на [близкие значения стандартных отклонений](#std) данных выборок, применить параметр, задающий, считать ли равными дисперсии генеральных совокупностей, из которых были сформированы выборки. Он задаётся по умолчанию, если не указать иного.

```python
results = st.ttest_ind(array1, array2, equal_var=True)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

- Однако даже при использовании этого параметра нулевая гипотеза отвержена. Принимаем альтернативную гипотезу **средняя выручка по всем периодам пользователей из Москвы отличается от выручки по всем периодам пользователей из других регионов**.

#### Проверка гипотезы 2 на выборке со средними значениями по пользователям

Заводим переменные с выборками для разных тарифов.

```python
array1 = total_money_for_users[total_money_for_users['city_category'] == 'Москва']['total_price']
array2 = total_money_for_users[total_money_for_users['city_category'] == 'другие']['total_price']
```

Проверка статистической гипотезы.

```python
results = st.ttest_ind(array1, array2)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

- При проверке нулевой гипотезы для средних значений всех пользователей не получилось отвергнуть нулевую гипотезу. Если учитывать для каждого пользователя среднюю выручку по всем периодам, а потом среднюю выручку от пользователей из Москвы сравнить со средней выручкой от пользователей из других городов, то **средняя выручка пользователей из Москвы равна выручке пользователей из других регионов**.
- Мы видим, что гипотезы при проверке на разных выборках дали разные результаты. Для второй гипотезы при первой проверке мы приняли альтернативную гипотезу о неравенстве выручек от пользователей из Москвы и из других регионов, однако вторая проверка не позволила отвергнуть нулевую гипотезу. Это может свидетельствовать о том, что, во-первых, при усреднении всех периодов и усреднении всех периодов сначала по пользователю результаты будут различны, и во-вторых, мы могли ошибиться при выборе критического уровня статистической значимости, попав в те самые 5%, когда мы отвергаем нулевую гипотезу, хотя она является верной.

**Гипотезы**

В рамках данной работы были проверены две гипотезы.
- **Гипотеза 1**: средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются.
- **Гипотеза 2**: средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

Для проверки была применена гипотеза о равенстве средних двух генеральных совокупностей при критическом уровне статистической значимости 5%.

Каждая из гипотез была проверена на двух выборках:
- на выборке, учитывающей каждый полный период или каждый период для отказавшегося от тарифа пользователя,
- а так же на выборке со средними значениями выручки от каждого пользователя.

При двух проверках нулевая гипотеза 1 была отвергнута в пользу альтернативной гипотезы: **средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются**. 

При двух проверках гипотезы 2 нулевая гипотеза была отвергнута при учете всех периодов в пользу альтернативной гипотезы **средняя выручка по всем периодам пользователей из Москвы отличается от выручки по всем периодам пользователей из других регионов**.

Однако при проверке гипотезы 2 для средних значений всех пользователей не получилось отвергнуть нулевую гипотезу **средняя выручка пользователей из Москвы равна выручке пользователей из других регионов**.
Это может свидетельствовать как о влиянии большего количества учитываемых периодов у пользователей из Москвы, как и о том, что мы могли при выборе критического уровня статистической значимости попасть в те самые 5%, когда мы отвергаем верную нулевую гипотезу.
